# Les suites récurrentes

## Etude d'une suite

Nous allons étudier la suite suivante

$\left\{
    \begin{array}{ll}
        u_{n+1} = 3\sqrt{u_n} + 4 \\
        u_0 = 0
    \end{array}
\right.
$

1. Conjecturer graphiquement des résultats sur le comportement global et asymptotique de $(u_n)$

2. Démontrer par récurrence que $0 \leq u_n \leq u_{n+1} \leq 16$

3. En utilisant le théorème de convergence monotone montrer que  $(u_n)$ converge vers L un réel.

Nous allons dans un premier temps étudier la fonction $C_f$ définie comme suit :

$f(x) = 3 \times \sqrt{x} + 4$ l'étude de cette fonction se fera sur son domaine de définition.


```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

def f(x):

    return 3*np.sqrt(x) + 4

#premier terme de la suite
u=[0]

# Nombre de points à représenter
n=16

for i in range(n):
    u.append(f(u[i]))

# coordonnées sur l'axe des abscisses
x=np.arange(0,16,0.01)

plt.plot(x, f(x), 'b-', x, x, 'r-')

# graphiquement
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


# on trace maintenant les traits noir k correspond à black

for i in range(0,n-1):
    plt.plot([u[i], u[i+1]], [u[i+1], u[i+1]], 'k-')
    plt.plot([u[i+1], u[i+1]], [u[i+1], u[i+2]], 'k-')

    plt.plot([u[i], u[i]], [0, u[i+1]], 'k--')

plt.plot([u[n-1], u[n-1]], [0, u[n]], 'k--')


plt.annotate('$u_0$', xy=(u[0], 0), xytext=(u[0]+.1, .1),

            arrowprops=dict(arrowstyle="->",

                            connectionstyle="arc3")

            )

plt.annotate('$u_1$', xy=(u[1], 0), xytext=(u[1]+.1, .1),

            arrowprops=dict(arrowstyle="->",

                            connectionstyle="arc3")

            )

# on met un titre, on peut taper en tex! 

plt.title('Etude de $u_{n+1}=f(u_n)$')

# on affiche le dessin

plt.show()
```

### corrections

D'après le graphique la suite est croissante et converge semble-t-il vers une valeur L = 16

2. Démonstration par récurrence

- Initialisation

$0 \leq u_0 \leq u_1 \leq 16$

$0 \leq 0 \leq 4 \leq 16$

Donc clairement cette dernière inégalité est vraie donc l'initialisation est vérifiée.

- Héridité

Supposons qu'au rang n que $P(n)$ soit vraie :
$0 \leq u_n \leq u_{n+1} \leq 16$

Essayons de démontre qu'alors $P(n+1)$ est vraie.

Dérivons la fonction f(x) est dérivable sur $]0,16]$ et vaut : $f(x) = \frac{3}{2\sqrt{x}} \geq 0$ pour toute valeur de x dans l'intervalle $]0,16]$. La fonction f est donc strictement croissante sur cette intervalle, on a donc :

$0 = f(0) \leq f(u_n) \leq f(u_{n+1}) \leq f(16) = 16$

donc $0 \leq u_{n+1} \leq u_{n+2} \leq 16$ et donc $P(n+1)$ est vraie, l'héridité est donc vérifiée.

Conslusion nous vennons de démontrer par récurrence que $P(n)$ est vraie pour tout $n \in [0,16]$.

3. Convergence

D'après ce qui précède la suite est croissante et majorée donc d'après le théorème de convergence monotone on peut affirmer que la suite $(u_n)$ est convergente vers une limite L.

Comme d'une part $u_{n+1}=f(u_n)$ et d'autre part f est continue et croissante sur l'intervalle $[0,16]$ d'autre part nous avons montrer que pour tout valeur n la suite est telle que $u_n \in [0,16]$. On en déduit d'après le théorème de convergence monotone que la suite converge vers une L que l'on détermine alors avec l'équation suivante : $f(L) = L$.

$f(x) = x$ 

$3\sqrt{x} +4 = x$

$9 \times x = (x - 4)^2$

$9x = x^2 - 8x + 16$

$9x = x^2 - 8x + 16$

$x^2 - 17x + 16 = 0$

En utilisant le discriminant on trouve deux valeurs $x_1 = 1$ et $x_2=16$ il est clair que $x_1$ n'a pas de sens on en déduit donc que la limite pour cette suite est $L=16$.
